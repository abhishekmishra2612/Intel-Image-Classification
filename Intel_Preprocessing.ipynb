```python
import warnings
warnings.filterwarnings('ignore')
import pandas as pd
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
import seaborn as sns
import os
import cv2
```

```python
Labels_dict = {0:'buildings', 1:'forest',2:'mountain',3:'sea', 4:'street'}
name_to_int = {'buildings':0, 'forest':1, 'mountain':2, 'sea':3, 'street':4}
```

## Load Data.
 - 990 for Training.
 - 827 for Validation.

```python
def load_data(folder):   # Function to load data from disk and append into list of images and labels.
    images = list()
    labels = list()
    for label in list(Labels_dict.values()):
        list_img = os.listdir(str(folder)+'/'+str(label))
        for file in list(list_img):
            img = cv2.imread(str(folder)+'/'+str(label)+'/'+str(file))
            img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
            images.append(img)
            labels.append(name_to_int[label])
    return np.array(images) , np.array(labels)       
```

```python
train_images, train_labels = load_data('intel_train') # train data in intel_train folder.
train_images.shape, train_labels.shape
```

```python
test_images, test_labels = load_data('intel_test') # test data in intel_test folder.
```

## Some of the images and their class label.

```python
for i in [100, 300, 500, 700, 900]:
    print("Class Label =", Labels_dict[train_labels[i]])
    plt.imshow(train_images[i])
    plt.axis('off')
    plt.show()
```

```python
# load data into drive for further use.
np.save('intel_train_images.npy', train_images)
np.save('intel_train_labels.npy', train_labels)
np.save('intel_test_images.npy', test_images)
np.save('intel_test_labels.npy', test_labels)
```
